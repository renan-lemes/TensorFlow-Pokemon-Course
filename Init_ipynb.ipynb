```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('pokemon_alopez247.csv')
```

```python
df
```

```python
df = df[['isLegendary','Generation', 'Type_1', 'Type_2', 'HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed','Color','Egg_Group_1','Height_m','Weight_kg','Body_Style']]
```

```python
df['isLegendary'] = df['isLegendary'].astype(int)
```

```python
def dummy_creation(df, dummy_creation):
    for i in dummy_creation:
        df_dummy = pd.get_dummies(df[i])
        df = pd.concat([df,df_dummy],axis=1)
        df = df.drop(i, axis=1)
    return(df)
```

```python
df = dummy_creation(df, ['Egg_Group_1', 'Body_Style', 'Color','Type_1', 'Type_2'])
```

```python
df
```

```python
def train_test_splitter(DataFrame, colunm):
    df_train = DataFrame.loc[df[colunm] != 1]
    df_test = DataFrame.loc[df[colunm] == 1]

    df_train = df_train.drop(colunm, axis=1)
    df_test = df_test.drop(colunm, axis=1)

    return(df_train, df_test)
    #Função que divide o data frame em dois de teste e de traino da geração  Red Blue e da Yellow serão nosso teste e a 
    # e as demais geração serão as de treino # 

df_train, df_test = train_test_splitter(df,'Generation')
```

```python
def label_delineator(df_train, df_test, label):
    train_data = df_train.drop(label,axis=1).values ##
    train_labels = df_train[label].values
    test_data = df_test.drop(label, axis=1).values
    test_labels = df_test[label].values
    # Essa função separa os rotulos como no nosso caso usaremos isLegendary 
    # Ele separa os dados em forma de matriz que é o valor que nosso algoritmo entende #

    return(train_data, train_labels, test_data, test_labels)
```

```python
train_data, train_labels, test_data, test_labels = label_delineator(df_train, df_test, 'isLegendary')
```

```python
def data_normalizer(train_data, test_data):
    train_data = preprocessing.MinMaxScaler().fit_transform(train_data)
    test_data = preprocessing.MinMaxScaler().fit_transform(test_data)
    # Função que normaliza os dados e transforma e deixa eles na mesma escala #

    return(train_data, test_data)
```

```python
train_data, test_data = data_normalizer(train_data, test_data)
```

```python
# Keras é um API para o tensorflow 
# um modelo é contruido sobre camadas nesse modelo vamos fazer duas camadas neurais  conectadas 
# o numero de camadas associados a eles serão o numero de camadas.
# Na primeira camada vai ser usado a função 'ReLU' (Rectified Linear Unit) e na primeira camada precisamos especificar nosso input_size
# a forma de entrada em nosso conjunto de dados
# Depois  é feito uma outra camada softmax que é um tipo de regressão logica feita para situações de multiplos casos
# Como no nosso caso é Lendario e não Lendario , com isso delinearemos as possiveis indentidades do pokémon em dois grupos de possibilidades#
```

```python
length = train_data.shape[1]

model = keras.Sequential()
model.add(keras.layers.Dense(500, activation='relu', input_shape=[length,]))
model.add(keras.layers.Dense(2, activation='softmax'))
```

```python
print(model)
```

```python
print(length)
```

```python
model.compile(optimizer = 'sgd', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
# aqui compilamos o modelo e alimetamos ele com tres parametro 
# optomozer que determina como o modelo é atualizado à medida que ganha informação 
# loss uma função de perda que mede a precisão do modelo durante o treinamento 
# metrics que especifica quais informações ele fornece para que possamos analisar o modelo 
# O otimizador que usamos foi Stocha Gradient Descent (SGD) 
# Para a perda o loss estamos usando  o sparse_categorical_crossentropy  se nossos valores tivessem codificação one-hot , era melhor usar  categorial_crossentropy#
```

```python
model.fit(train_data, train_labels, epochs=400)
```
